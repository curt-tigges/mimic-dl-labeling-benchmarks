### HLAN Data Preparation

In order to process files for the HLAN model training, simply put the files that result from the CAML-MIMIC data preprocessing through this pipeline. To train the model, you will need dev_50.csv, test_50.csv and train_50.csv.

```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv('dev_50.csv')
```

```python
df
```

```python
for i in range(len(df)):
    df.iloc[i, 3] = '__label__' + df.iloc[i, 3]
```

```python
df
```

```python
for i in range(len(df)):
    df.iloc[i, 3] = df.iloc[i, 3].replace(';',' ')
```

```python
df
```

```python
df = df.drop(['SUBJECT_ID', 'HADM_ID', 'length'], axis=1)
```

```python
df
```

```python
with open('mimiciii_dev_50_th0.txt', 'w+') as f:
    for i in range(len(df)):
    #train_50.iloc[i, 3] = train_50.iloc[i, 3].replace(';',' ')
        line = df.iloc[i,0] + ' ' + df.iloc[i, 1] + '\n'
        f.write(line)
```
